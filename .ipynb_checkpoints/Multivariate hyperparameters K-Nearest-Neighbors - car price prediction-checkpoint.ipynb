```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
import re
```

```python
# Source: https://archive.ics.uci.edu/ml/datasets/Automobile
cars = pd.read_csv("imports-85.data", header=None)
```

```python
cars.head()
```

```python
cols = """1. symboling: -3, -2, -1, 0, 1, 2, 3.
2. normalized-losses: continuous from 65 to 256.
3. make:
alfa-romero, audi, bmw, chevrolet, dodge, honda,
isuzu, jaguar, mazda, mercedes-benz, mercury,
mitsubishi, nissan, peugot, plymouth, porsche,
renault, saab, subaru, toyota, volkswagen, volvo

4. fuel-type: diesel, gas.
5. aspiration: std, turbo.
6. num-of-doors: four, two.
7. body-style: hardtop, wagon, sedan, hatchback, convertible.
8. drive-wheels: 4wd, fwd, rwd.
9. engine-location: front, rear.
10. wheel-base: continuous from 86.6 120.9.
11. length: continuous from 141.1 to 208.1.
12. width: continuous from 60.3 to 72.3.
13. height: continuous from 47.8 to 59.8.
14. curb-weight: continuous from 1488 to 4066.
15. engine-type: dohc, dohcv, l, ohc, ohcf, ohcv, rotor.
16. num-of-cylinders: eight, five, four, six, three, twelve, two.
17. engine-size: continuous from 61 to 326.
18. fuel-system: 1bbl, 2bbl, 4bbl, idi, mfi, mpfi, spdi, spfi.
19. bore: continuous from 2.54 to 3.94.
20. stroke: continuous from 2.07 to 4.17.
21. compression-ratio: continuous from 7 to 23.
22. horsepower: continuous from 48 to 288.
23. peak-rpm: continuous from 4150 to 6600.
24. city-mpg: continuous from 13 to 49.
25. highway-mpg: continuous from 16 to 54.
26. price: continuous from 5118 to 45400."""

cols = re.sub("\\n","",cols)
cols = re.findall("(\d+\.\s.+?:)",cols)
cols = [re.sub("(\d*\.\s)","",col) for col in cols]
cols = [re.sub(":","",col) for col in cols]
cols
```

```python
cars.columns = cols
cars.head(10)
```

```python
(cars == "?").sum().sort_values(ascending=False)
```

```python
cars[["normalized-losses", "price", "stroke", "bore", "peak-rpm", "horsepower", "num-of-doors"]].head(10)
```

```python
print(cars["num-of-doors"].value_counts())
cars["num-of-doors"] = cars["num-of-doors"].replace("four",4).replace("two",2)
```

Replaced car doors to numerical values

```python
cars.info()
```

```python
cars_num = cars[["normalized-losses", "wheel-base", "length", "width", "height", "curb-weight", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]].replace("?", np.nan)
cars_num = cars_num.astype(float)
```

Replaced all null values of numerical cars columns with np.nan and changed its dtype to float

```python
cars_num.dropna(subset=["price"])
cars_num = cars_num.fillna(cars_num.mean())
```

Dropped all rows with missing 'price' column since it is the target column

```python
cars_num.mean()
```

After, we will 'minmax normalize' all numerical column (except 'price' column) with (val - min)/(max - min)

```python
temp = cars_num["price"]
cars_norm = (cars_num - cars_num.min()) / (cars_num.max() - cars_num.min())
cars_norm["price"] = temp
cars_norm.head(10)
```

```python
def knn_train_test(tra_col, tar_col, df, k=5):
    np.random.seed(1)
    df = df.loc[np.random.permutation(len(df))]
    border = int(0.8 * len(df))
    train = df[:border]
    test = df[border:]
    knr = KNeighborsRegressor(n_neighbors=k)
    knr.fit(train[tra_col], train[tar_col])
    pred = knr.predict(test[tra_col])
    rmse = mean_squared_error(test[tar_col], pred)**0.5
    return rmse
```

Try out every K-Nearest Neighbors on every column with k=5 hyperparameter.

```python
for col in cars_norm.columns[cars_norm.columns != "price"]:
    print(col, knn_train_test([col], ["price"], cars_norm))
```

Lowest RMSE using k=5 for univariate K Nearest Neighbors is the "curb-weight" column. Next, we'll try different k values of 1,3,5,7,9 and trying different columns again.

```python
k_vals = [1,3,5,7,9]
k_rmse = {}

for col in cars_norm.columns[cars_norm.columns != "price"]:
    k_rmse[col] = {}
    for val in k_vals:
        k_rmse[col][val] = knn_train_test(k=val, tra_col=[col], tar_col=["price"], df=cars_norm)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
rmse_df = pd.DataFrame(k_rmse)

plt.figure(figsize=(10,10))
plt.ylabel("RMSE")
plt.xlabel("k value")
plt.xticks([1,3,5,7,9])

for col in rmse_df.columns:
    rmse_df[col].plot(legend=True)
    
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

Lowest RMSE with lowest k-value is curb-weight column with k-value of 5.

```python
asc_rmse = rmse_df.loc[5].sort_values()
top_5_rmse = asc_rmse[:5]
print("Top 5 variables with k=5:")
top_5_rmse
```

```python
multivariate_rmse = {}
mult_cols = list(top_5_rmse.index)

for mul in range(2,6):
    #print(mult_cols[:mul])
    multivariate_rmse[mul] = knn_train_test(k=5, tra_col=mult_cols[:mul], tar_col=["price"], df=cars_norm)
    
pd.Series(multivariate_rmse).sort_values()
```

* Top 4 combined variables of k=5 had RMSE of 3022
* Top 3 combined variables of k=5 had RMSE of 3226
* Top 5 combined variables of k=5 had RMSE of 3367
* Top 2 combined variable of k=5 had RMSE of 3460

So, combination of top 4 variables (curb-weight, highway-mp, city-mpg, length) with k-value of 5 gave the lowest RMSE.

```python
top_3_models = pd.Series(multivariate_rmse).sort_values()[:3].index
hypers = list(range(1,26))
multivar_k_rmse = {}

#loop over number of top varibles
for mul in top_3_models:
    multivar_k_rmse[mul] = {}
    #loop over k-value of 1 to 26
    for hyp in hypers:
        multivar_k_rmse[mul][hyp] = knn_train_test(k=hyp, tra_col=mult_cols[:mul], tar_col=["price"], df=cars_norm)
```

```python
multivar_k_df = pd.DataFrame(multivar_k_rmse)
multivar_k_df.index.name = "k-values"
multivar_k_df.columns = ["top 3 vars", "top 4 vars", "top 5 vars"]
```

```python
multivar_k_df
```

```python
for i in multivar_k_df.columns:
    print(i,":", multivar_k_df[i].idxmin(), "k-values. RMSE:", round(multivar_k_df[i].min(), 2))
```

```python
plt.figure(figsize=(10,10))
plt.ylabel("RMSE")
for col in multivar_k_df.columns:
    multivar_k_df[col].plot(legend=True)
    
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

After exploring 26 different k-values and combinations of top 3 to top 5 different variables, the lowest RMSE of 2646 was achieved by K-Nearest Neighbor of k=4 and combination of top 3 variables of curb-weight, highway-mp, and city-mpg
